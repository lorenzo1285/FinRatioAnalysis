# Package Import

```python
import pandas as pd
from FinRatioAnalysis import FinRatioAnalysis
```

```python
tickers = ['IBM', 'AAPL', 'GOOG']
```

```python
objetos = {}

for ticker in tickers:
    obj = FinRatioAnalysis(ticker, 'yearly')
    objetos[ticker] = obj

# Ahora objetos es un diccionario con claves 'IBM', 'AAPL' y 'GOOG' y valores que son objetos de la clase FinRatioAnalysis

# Acceder al objeto para el ticker 'IBM'
obj_ibm = objetos['IBM']

# Acceder al objeto para el ticker 'AAPL'
obj_aapl = objetos['AAPL']

# Acceder al objeto para el ticker 'GOOG'
obj_goog = objetos['GOOG']
```

## New Feature: Valuation & Growth Metrics

The `valuation_growth_metrics()` method provides hybrid metrics for asset allocation:
- **Valuation metrics**: P/E, P/S, dividend yield, payout ratio
- **Growth metrics**: 3-year CAGR for revenue and net income

```python
# Get valuation and growth metrics for IBM
obj_ibm.valuation_growth_metrics()
```

```python
# Compare metrics across multiple companies
for ticker, obj in objetos.items():
    print(f"\n{ticker} Valuation & Growth Metrics:")
    display(obj.valuation_growth_metrics())
```
